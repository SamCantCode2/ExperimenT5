```python
sentences = [
    'Ich habe keine lust',
    'Ich will eure hande sehen',
    'Ich will das ihr mir vertraut',
    'Ich hasse kinder',
    'Ich bin schon langen kallt'
]
```

```python
tk = Tokenizer(num_words = 200)
tk.fit_on_texts(sentences)
wordindex = tk.word_index
wordindex
```

```python
sequences = tk.texts_to_sequences(sentences)
sequences
```

Testing with unknown sequences to check for tokenization

```python
testsentences = [
    'Ich will eure phantasie',
    'Ich liebe dich',
    'Nein ich hasse lust'
]
```

```python
testseq = tk.texts_to_sequences(testsentences)
testseq
```

```python
tk2 = Tokenizer(num_words = 200, oov_token = "{bro idk}")
tk2.fit_on_texts(sentences)
wordindex = tk2.word_index
wordindex
```

```python
testseq = tk2.texts_to_sequences(testsentences)
testseq
```

Testing sequences with padding to check if it maintains equal length

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded = pad_sequences(testseq, maxlen=5)
padded
```

```python
padded = pad_sequences(testseq, padding='post')
padded
```

```python
padded = pad_sequences(testseq, maxlen=3, truncating='post')
padded
```

```python
padded = pad_sequences(testseq, maxlen=3, truncating='pre')
padded
```
